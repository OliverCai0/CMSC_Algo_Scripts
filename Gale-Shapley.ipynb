```python
def gs(pref):
    s = {}
    free = list(pref['A'].keys())
    propose = dict([(m,list(pref['A'][m])) for m in free])
    step = 1
    while free:
#         print(f'Step {step} \\\\')
        m = free.pop(0)
        print(f'\item Selected free {m}, who has not proposed to {propose[m]} \\\\')
        w = propose[m].pop(0)
        print(f'Selected highest preference {w}\\\\')
        if w in s.keys():
            if pref['B'][w].index(m) < pref['B'][w].index(s[w]):
                print(f'{m} ranked higher on {w}\'s list, so pair ({s[w]}, {w}) is replaced by ({m},{w}) \\\\')
                free.append(s[w])
                s[w] = m
            else:
                free.append(m)
                print(f'{m} ranked lower on {w}\'s list compared to {s[w]}, so {m} is free \\\\')
        else:
            print(f'{w} not proposed to yet, so pair ({m}, {w}) added to S \\\\')
            s[w] = m
        step += 1
    return dict([(s[w],w) for w in s.keys()])
```

```python
def isStable(match, prefA, prefB):
    printout = ''
    for key in m.keys():
        printout += '('+ key + ',' + m[key] + '),' 
    print('\item Is \{ ' + printout + ' \} stable? \\\\')
    for pair in match.keys():
        #put together list of potential cheating partners
        cheaters = []
        potential = prefA[pair][:(prefA[pair].index(match[pair]))]
        print(f'Current partner of {pair} is {match[pair]} \\\\')
        print(f'Potential cheating partners for {pair}: {potential} \\\\')
        for c in potential:
            current_partner = list(match.keys())[(list(match.values()).index(c))]
            print(f'Current partner of {c} is {current_partner} \\\\')
            #is pair[0] higher than c's current partner on their pref list?
            if prefB[c].index(pair) < prefB[c].index(current_partner):
                print(f'{pair} is higher than {current_partner} on {c}\'s preference list \\\\')
                print(f'{[(match[m],match[m]) for m in match.keys()]} is not stable')
                return False
            print(f'{pair} is lower than {current_partner} on {c}\'s preference list \\\\')
    return True
        
```

```python
#pairing = [actual_pairing,remaining_A,remaining_B]
def createPossiblePairings(pairings):
    collection = []
    while pairings:
        p = pairings.pop(0)
        A_member = p[1].pop(0)
        B = p[2]
        #print(B)
        for index in range(len(B)):
            actual_pairing = p[0] + [[A_member, B[index]]]
            if len(p[1]) > 0:
                #print(f'Added {actual_pairing}')
                #print(f'New object{[actual_pairing, p[1], B[:index] + B[index + 1:]]}')
                pairings.append([actual_pairing, [x for x in p[1]], B[:index] + B[index + 1:]])
            else:
                collection.append(actual_pairing)
    return collection
            
                   
```
